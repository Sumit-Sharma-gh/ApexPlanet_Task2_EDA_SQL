## SQL PART

```python
import sqlite3
```

```python
df.info()
```

```python
conn = sqlite3.connect(':memory:')
df.to_sql('sales', conn, index=False, if_exists='replace')
```

```python
pd.read_sql("""
select distinct count("Order ID") as "Total Orders" from sales; """, conn)
```

## Top 5 Product Name with most Sales

```python

pd.read_sql("""
SELECT "Product Name", SUM(Sales) AS Total_Sales
FROM sales
GROUP BY "Product Name"
ORDER BY Total_Sales DESC
LIMIT 5;
""", conn)
# The highest Revenue done by "Canon imageCLASS 2200 Advanced Copier" product
```

## Region-wise Total Sales


```python
pd.read_sql("""
select Region , sum(Sales) as Total_Sales
from sales
group by Region
order by Total_Sales desc
;""",conn)

# West region Contributes the highest Revenue and South Region Contributes lowest Revenue
```

```python
pd.options.display.float_format = '{:,.2f}'.format
```

## Segment-wise Revenue


```python
pd.read_sql(""" 
select Segment, sum(Sales) as Total_Sales
from sales 
group by Segment
order by Total_Sales
desc;

""", conn)

#The Consumer segment generates the highest revenue compared to Corporate and Home Office segments
```

## Monthly Sales Trend


```python
pd.read_sql("""
select "Order Month", sum(Sales) as Total_Sales
from sales
group by "Order Month";
""", conn)

# Monthly sales show fluctuations throughout the year, indicating potential seasonal patterns in customer purchasing behavior.
```

## Average Shipping Days by Region


```python
pd.read_sql(""" 
select Region, Round(avg("Shipping Days"),2) as Avg_Shipping_Days 
from sales
group by Region;
""", conn)

#Average shipping time differs across regions, which may indicate variations in logistics efficiency or geographical delivery challenges.
```

## Top 5 Cities by Sales


```python
pd.read_sql(""" 
select City, sum(Sales) as Total_Sales
from sales
group by City
order by Total_Sales
desc limit 5;
""",conn)

#A few major cities contribute disproportionately to total revenue, indicating concentrated customer demand in key urban markets.
```

##  Category-wise Average Sales


```python
pd.read_sql(""" 
select Category , sum(Sales) as Total_Sales
from sales
group by Category
""",conn)

#Technology Category Item Contributes the most number of Sales as Compared to Other Category
```

## Orders Count by Segment


```python
pd.read_sql("""

select Segment, Count("Order ID") as Total_Orders
from sales
group by Segment;
""",conn)

#This Shows the total Number of order happened in each Segements
```

## Year-wise Total Sales


```python
pd.read_sql("""
select "Order Year", Sum(Sales) as Total_Sales
from sales
group by "Order Year";
""",conn)

#Sales demonstrate variation across years, which may indicate business growth trends or changes in market demand over time.
```

## High Value(Sale > 1000) Orders Count


```python
pd.read_sql("""
select count("Order ID") as Total__Orders
from sales
where Sales > 1000;
""",conn)

#High-value transactions represent a smaller proportion of total orders but contribute significantly to overall revenue 
#highlighting the importance of premium purchases.
```
